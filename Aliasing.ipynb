```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt
from numpy import pi
```

```python
n_axis = np.arange(8) # x-ova osa pro diskretni prubeh n=0..7
x_axis = np.linspace(0,8,100) # x-ova osa pro pomocny spojity prubeh

# cosinus na frekvenci 1/8
cos_n1 = np.cos(2*pi*2/8*n_axis) # diskretni
cos_x1 = np.cos(2*pi*2/8*x_axis) # spojity

# cosinus na frekvenci 7/8
cos_n2 = np.cos(2*pi*6/8*n_axis) # diskretni
cos_x2 = np.cos(2*pi*6/8*x_axis) # spojity

# vykresleni obou
plt.figure()
plt.stem(n_axis, cos_n1)
plt.plot(x_axis, cos_x1,color='0.5')

plt.figure()
plt.stem(n_axis, cos_n2)
plt.plot(x_axis, cos_x2,color='0.5')

plt.show()
```
