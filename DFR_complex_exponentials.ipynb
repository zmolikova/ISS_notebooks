```python
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d.art3d as art3d
```

```python
t_axis = np.linspace(0,4,1000)
n_axis = np.arange(4)
```

```python
f = plt.figure(figsize=(8,12))

for i in range(4):
    # spojita komplexni exponenciala
    exp_pi2_cont = np.cos(-i*pi/2 * t_axis) + 1j * np.sin(-i*pi/2 * t_axis)
    # diskretni komplexni exponenciala
    exp_pi2_disc = np.cos(-i*pi/2 * n_axis) + 1j * np.sin(-i*pi/2 * n_axis)
    
    # vykresleni spojite exponencialy
    ax = f.add_subplot(4,1,i+1, projection='3d')
    ax.plot(t_axis, np.real(exp_pi2_cont), np.imag(exp_pi2_cont))
    
    # vykresleni x-ove osy
    line = art3d.Line3D((0,4),(0,0),(0,0), color='green')
    ax.add_line(line)
    
    # vykresleni diskretni exponencialy
    for x,y,z in zip(n_axis, np.real(exp_pi2_disc), np.imag(exp_pi2_disc)):
        line = art3d.Line3D(*zip((x,0,0), (x,y,z)), marker='o', markevery=(1, 1), color='red')
        ax.add_line(line)
    
    ax.set_xlabel('t')
    ax.set_ylabel('$\mathbb{R}$')
    ax.set_zlabel('$\mathbb{I}$')
    ax.set_xlim([0,4])
    ax.set_ylim([-1.5, 1.5])
    ax.set_zlim([-1.5, 1.5])
plt.show()
```
