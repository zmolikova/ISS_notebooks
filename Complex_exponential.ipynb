```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi as pi
from mpl_toolkits.mplot3d import Axes3D

from IPython.display import display
from ipywidgets import widgets
```

```python
fig = plt.figure(figsize=(9,4))

ax1 = fig.add_subplot(1, 2, 1)
# unit circle
ang = np.linspace(0,2*np.pi,1000)
ax1.plot(np.cos(ang), np.sin(ang))
ax1.set_xlabel('$\mathbb{R}\{e^{jx}\}$')
ax1.set_ylabel('$\mathbb{I}\{e^{jx}\}$')
# dot showing current position on unit circle
point_plot, = ax1.plot([1],[0],'o', markersize = 8)

ax2 = fig.add_subplot(1,2,2, projection='3d')
# complex exponential
t = np.linspace(0,4*pi,1000)
r = np.cos(t)
i = np.sin(t)
ax2.plot(t,r,i)
ax2.set_xlabel('$x$')
ax2.set_ylabel('$\mathbb{R}\{e^{jx}\}$')
ax2.set_zlabel('$\mathbb{I}\{e^{jx}\}$')
# dot showing current position on complex exponential
point_plot2, = ax2.plot([0],[1],[0],'o',markersize=8)

# wrapper for interactive mode
def plot_point(x=0.0):
    point_plot.set_data([np.cos(x)], [np.sin(x)])
    point_plot2.set_data([x], [np.cos(x)])
    point_plot2.set_3d_properties([np.sin(x)])
    plt.draw()

widgets.interact(plot_point, 
                 x = (0,4*pi, pi/64))
    
plt.show()
```
